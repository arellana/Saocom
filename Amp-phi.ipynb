```python
HHhist = HH.flatten()
HVhist = HV.flatten()
VHhist = VH.flatten()
VVhist = VV.flatten()

hist = [HHhist,HVhist,VHhist,VVhist]
hist = np.asarray(hist)
hist.shape
histdb = np.zeros((4,len(HHhist)))
histdb.shape
```

```python
def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
med = []
desv = []

for i in range(len(hist)):
    # i=0
    plt.figure(figsize=(5,5))

    histdb[i] = 10*np.log10(hist[i]) 
    n, b= np.histogram(histdb[i], 100)

    bin_max = np.where(n == n.max())
    cota = int(len(n)/1)

    x = b[:-1]
    y = n[:]

    popt, pcov = curve_fit(gaus,x,y,maxfev = 1000000, p0=[250,-20,8])
    med.append(popt[1])
    desv.append(popt[2])

    # print('media:', med, '\n'+'desviacion: ,', desv)
    # print(np.sqrt(np.diag(pcov))[1:],'\n')
    #Grafico solo HH y VV
    if (i == 0 or i == 3): 
        n, bins, patches = plt.hist(histdb[i], 100, facecolor='grey', alpha=0.75, label = 'Datos')#, density=True, stacked=True)
        plt.plot(x,gaus(x,*popt),'b',label='Distribucin normal', linewidth = 2)


    # n, bins, patches = plt.hist(histdb[i], 100, facecolor='grey', alpha=0.75, label = 'Datos')#, density=True, stacked=True)
    # plt.plot(x,gaus(x,*popt),'b',label='Distribucin normal', linewidth = 2)

print('angInc: ', np.mean(angInc))    
print('Media HH-VV: ', [med[0],med[3]],'\n'+'Desv HH-VV',[desv[0],desv[3]])
```

```python
# altC,ancC = 3,4 #Le resto los ultimos datos para que quede multiplos de 10
altC,ancC = (HH.shape[0]%10), (HH.shape[1]%10)

if (altC == 0):
    HH2 = HH[:,:-ancC]
    VV2 = VV[:,:-ancC]
    ang2 = angInc[:,:-ancC]
elif (ancC == 0):
    HH2 = HH[:-altC,:]
    VV2 = VV[:-altC,:]
    ang2 = angInc[:-altC,:]
else:
    HH2 = HH[:-altC,:-ancC]
    VV2 = VV[:-altC,:-ancC]
    ang2 = angInc[:-altC,:-ancC]
    
print(HH2.shape)

def split(array, ncols, nrows):
    """Split a matrix into sub-matrices."""

    r, h = array.shape
    return (array.reshape(h//nrows, nrows, -1, ncols)
                 .swapaxes(1, 2)
                 .reshape(-1, nrows, ncols))


reescala = 2
splHH = split(HH2,int(len(HH2[:,0])/10)*reescala,int(len(HH2[0,:])/10)*reescala)
splVV = split(VV2,int(len(VV2[:,0])/10)*reescala,int(len(VV2[0,:])/10)*reescala)
splAng = split(ang2,int(len(ang2[:,0])/10)*reescala,int(len(ang2[0,:])/10)*reescala)
# splHH = split(HH2,16,22)
# splVV = split(VV2,10,10)
splHH.shape
```

```python
medHH = []
desvHH = []
medVV = []
desvVV = []
angg = []
bines = 10

plt.figure(figsize=(5,5))
for j in range(len(splHH[:,0,0])):

    histhh = splHH[j].flatten()
    histhhdb = 10*np.log10(histhh)
    n, b= np.histogram(histhhdb, bines)

    bin_max = np.where(n == n.max())

    x = b[:-1]
    y = n[:]

    popt, pcov = curve_fit(gaus,x,y,maxfev = 1000000, p0=[5,-14,2])
    medHH.append(popt[1])
    desvHH.append(popt[2])
    # plt.figure(figsize=(3,3))
    plt.hist(histhhdb, bines, facecolor='grey', alpha=0.75, label = 'Datos')#, density=True, stacked=True)
    plt.plot(x,gaus(x,*popt),'b',label='Distribucin normal', linewidth = 2)
    
    
    histvv = splVV[j].flatten()
    histvvdb = 10*np.log10(histvv)
    n, b= np.histogram(histvvdb, bines)

    bin_max = np.where(n == n.max())

    x = b[:-1]
    y = n[:]

    popt, pcov = curve_fit(gaus,x,y,maxfev = 1000000, p0=[10,-16,1])
    medVV.append(popt[1])
    desvVV.append(popt[2])
    
    # plt.figure(figsize=(3,3))    
    plt.hist(histvvdb, bines, facecolor='grey', alpha=0.75, label = 'Datos')#, density=True, stacked=True)
    plt.plot(x,gaus(x,*popt),'b',label='Distribucin normal', linewidth = 2)
    
    angg.append(np.mean(splAng[j]))
```
